```python
min_val = crime['TOTAL IPC CRIMES'].min()
max_val = crime['TOTAL IPC CRIMES'].max()
range_val = (max_val - min_val) / 4
low = min_val + range_val
medium = low + range_val
high = medium + range_val
```

```python
def get_crime_level(crime_count):
    if crime_count <= low:
        return 1
    elif crime_count <= medium:
        return 2
    elif crime_count <= high:
        return 3
    else:
        return 4

crime['CRIME_LEVEL'] = crime['TOTAL IPC CRIMES'].apply(get_crime_level)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
file_path =  r"C:\Users\Heman\OneDrive\Desktop\crime (1).csv"  # Update with the correct path to your CSV file
crime_data = pd.read_csv(file_path)

# Aggregate the data by state
statewise_crime_data = crime_data.groupby('STATE/UT').sum(numeric_only=True).reset_index()

# Select relevant features and the target variable
features = statewise_crime_data.drop(columns=['STATE/UT', 'YEAR', 'TOTAL IPC CRIMES'])
target = statewise_crime_data['TOTAL IPC CRIMES']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set and evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')

# Function to predict the crime rate for a given state and return as a percentage
def predict_crime_rate(state_name):
    # Check if the state exists in the dataset
    if state_name not in statewise_crime_data['STATE/UT'].values:
        return f"State '{state_name}' not found in the dataset."
    
    # Get the features for the specified state
    state_features = statewise_crime_data[statewise_crime_data['STATE/UT'] == state_name].drop(columns=['STATE/UT', 'YEAR', 'TOTAL IPC CRIMES'])
    
    # Predict the total crime rate using the trained model
    predicted_crime_rate = model.predict(state_features)[0]
    
    # Calculate the percentage based on the maximum total crime rate in the dataset
    max_crime_rate = statewise_crime_data['TOTAL IPC CRIMES'].max()
    crime_rate_percentage = (predicted_crime_rate / max_crime_rate) * 100
    
    return f"The predicted total crime rate for {state_name} is {crime_rate_percentage:.2f}% of the maximum recorded crime rate."

# Example usage
example_state = "ANDHRA PRADESH"
print(predict_crime_rate(example_state))
```
